```python
cursor.execute("CREATE TABLE rest ("
               "id VARCHAR(255),"
               "x INT(255),"
               "y INT(255),"
               "tag VARCHAR(255),"
               "name VARCHAR(255),"
               "image VARCHAR(1023) )"
)

cnxn.commit()  # this commits changes to the database
```

```python
area = 'Gangnam'
rest = pd.read_csv(f'../crawings/area_csv/{area}/rest_concat.csv')

rest['id'] = rest['url'].astype('str')
rest['x'] = 0
rest['y'] = 0
rest['tag'] = '카페'
rest['name'] = rest['restaurant']
rest['image'] = 'https://search.pstatic.net/common/?autoRotate=true&type=w560_sharpen&src=https%3A%2F%2Fldb-phinf.pstatic.net%2F20210503_64%2F1619971687675Vy8gN_JPEG%2F7l3e4V0VZGXCM52wtpkqTPpe.jpg'
rest = rest[['id','x','y','tag','name','image']]
```

```python
query = "INSERT INTO rest (id, x, y, tag, name, image) VALUES (%s, %s, %s, %s, %s, %s)"
a = rest.values.tolist()
for i in range(len(a)):
    a[i] = tuple(list(a[i]))
```

```python
# 3000개 넣는데 6초 걸림
from tqdm import tqdm
for i in tqdm(a):
    cursor.executemany(query, [i])
cnxn.commit()
```
